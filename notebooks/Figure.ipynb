```python
import numpy as np
import matplotlib
from matplotlib.patches import Circle, Wedge
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig_width= 15
num = 6
r_cir = 0.1
s_arrow = 0.06
alpha = .4

c_f1 = 'k'
c_f2 = 'r'
c_cir = 'k'


fig, ax = plt.subplots(1, 1, figsize=(fig_width, fig_width/1.6180), sharex=True)
patches = []
patches2 = []

for x in range(int(num*1.6180)+1):
    if x < num :
        lala = range(x+1)
        lalay = num-1
    else:
        lala = range(num)
        lalay = lala[-1]
    for y in lala :
        if y < 5 and x < int(num*1.6180):
            patches += [Wedge((x,y), r_cir, 0,360, width=r_cir/10)]
        else :
            patches2 += [Wedge((x,y), r_cir, 0,360, width=r_cir/10)]
            
        if x < int(num*1.6180):
            if y <= lalay-1 :
                if y != lalay-1 and x < int(num*1.6180)-1 :
                    plt.arrow(x+r_cir, y+r_cir, 1-(r_cir*2)-(s_arrow/2), 1-(r_cir*2)-(s_arrow/2),
                              head_width=s_arrow, head_length=s_arrow, fc=c_f1, ec=c_f1)
                else :
                    plt.arrow(x+r_cir, y+r_cir, 1-(r_cir*2)-(s_arrow/2), 1-(r_cir*2)-(s_arrow/2),
                              head_width=s_arrow, head_length=s_arrow, fc=c_f1, ec=c_f1, alpha=alpha)
                
                if x < int(num*1.6180)-1 :
                    if y == 0 :
                        plt.arrow(x+r_cir, y, 1-(r_cir*2), -y, head_width=0., head_length=0., ec=c_f2)
                    else :
                        plt.arrow(x+r_cir, y-r_cir, 1-(r_cir*2), -y+r_cir,
                                  head_width=0.,head_length=0., ec=c_f2)
                else :
                    if y == 0 :
                        plt.arrow(x+r_cir, y, 1-(r_cir*2), -y,
                                  head_width=0., head_length=0., ec=c_f2, alpha=alpha)
                    else :
                        plt.arrow(x+r_cir, y-r_cir, 1-(r_cir*2), -y+0.05,
                                  head_width=0., head_length=0., ec=c_f2, alpha=alpha)
                
            else :
                plt.arrow(x+r_cir, y-r_cir, 1-(r_cir*2), -y+r_cir, head_width=0.,
                          head_length=0., ec=c_f2, alpha=alpha)

p = PatchCollection(patches, color=c_cir, lw=1)
ax.add_collection(p)
p = PatchCollection(patches2, color=c_cir,alpha=alpha, lw=1)
ax.add_collection(p)


plt.plot([0,1,2,3,0,0,1,2,0,1], lw=10, alpha=alpha, c='k')
plt.axis([-0.32,(num+0.32)*1.6180,-0.32,num+0.32])
plt.show()
```
